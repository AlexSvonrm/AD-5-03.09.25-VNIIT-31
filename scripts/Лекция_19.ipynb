### 1. Простое линейное график

```python
import matplotlib.pyplot as plt

'''Данные'''
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

'''Создание линейного графика'''
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Линейный график')
plt.title('Простой линейный график')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

'''Данные'''
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

'''Настройка графика'''
plt.figure(figsize=(12, 6))

'''Первый график: синус'''
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue', label='sin(x)', linewidth=2)
plt.title('График sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()

'''Второй график: косинус'''
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='orange', label='cos(x)', linewidth=2)
plt.title('График cos(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()

'''Третий график: тангенс'''
plt.subplot(2, 1, 2)
plt.plot(x, y3, color='green', label='tan(x)', linewidth=2)
plt.ylim(-10, 10)
plt.title('График tan(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()

'''Добавление аннотаций'''
plt.annotate('Местоположение максимума', xy=(1.57, 10), xytext=(2, 8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

'''Показать график'''
plt.tight_layout()
plt.show()
```

### 3. Круговая диаграмма

```python
'''Данные'''
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

'''Создание круговой диаграммы'''
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Круговая диаграмма')
plt.axis('equal')  # Чтобы круг выглядел как круг
plt.show()
```

```python
import matplotlib.pyplot as plt

'''Данные'''
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)

'''Создание круговой диаграммы'''
fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%',
                                   startangle=90, colors=colors, explode=explode)

'''Настройка аннотаций'''
for text in autotexts:
    text.set_color('white')
    text.set_fontsize(12)

'''Настройка заголовков'''
ax.set_title('Сложная круговая диаграмма', fontsize=16)

'''Добавление легенды'''
ax.legend(wedges, labels, title="Категории", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

'''Убедитесь, что круг выглядит как круг'''
plt.axis('equal')

'''Показать график'''
plt.tight_layout()
plt.show()
```

### 4. График рассеяния

```python
import numpy as np

'''Генерация случайных данных'''
x = np.random.rand(50)
y = np.random.rand(50)

'''Создание графика рассеяния'''
plt.scatter(x, y, color='purple', alpha=0.5)
plt.title('График рассеяния')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

'''Генерация случайных данных'''
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(50, 200, size=50)
colors = np.random.rand(50)

'''Создание графика рассеяния'''
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis', edgecolors='w')

'''Настройка заголовков и меток'''
ax.set_title('Сложный график рассеяния', fontsize=16)
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)

'''Добавление сетки'''
ax.grid(True)

'''Добавление цветовой шкалы'''
cbar = plt.colorbar(scatter)
cbar.set_label('Цветовая шкала', fontsize=12)

'''Аннотации для некоторых точек'''
for i in range(5):
    ax.annotate(f'Точка {i+1}', (x[i], y[i]), textcoords="offset points",
                xytext=(0,10), ha='center', fontsize=10, color='black')

'''Показать график'''
plt.tight_layout()
plt.show()
```

### 5. Создание нескольких графиков в одном окне

```python
'''Данные'''
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

'''Создание Figure и Axes'''
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

'''Первое подграфик'''
ax[0].plot(x, y1, color='blue', label='sin(x)')
ax[0].set_title('График sin(x)')
ax[0].legend()
ax[0].grid()

'''Второе подграфик'''
ax[1].plot(x, y2, color='red', label='cos(x)')
ax[1].set_title('График cos(x)')
ax[1].legend()
ax[1].grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

'''Данные'''
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

'''Создание Figure и Axes'''
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

'''Первое подграфик'''
ax[0].plot(x, y1, color='blue', label='sin(x)', linestyle='-', linewidth=2)
ax[0].set_title('График sin(x)', fontsize=16)
ax[0].set_xlabel('X', fontsize=12)
ax[0].set_ylabel('sin(x)', fontsize=12)
ax[0].legend()
ax[0].grid(True)

'''Аннотация для максимума и минимума'''
ax[0].annotate('Максимум', xy=(1.57, 1), xytext=(2, 0.5),
                arrowprops=dict(facecolor='black', shrink=0.05),
                fontsize=10, color='blue')
ax[0].annotate('Минимум', xy=(4.71, -1), xytext=(5, -0.5),
                arrowprops=dict(facecolor='black', shrink=0.05),
                fontsize=10, color='blue')

'''Второе подграфик'''
ax[1].plot(x, y2, color='red', label='cos(x)', linestyle='--', linewidth=2)
ax[1].set_title('График cos(x)', fontsize=16)
ax[1].set_xlabel('X', fontsize=12)
ax[1].set_ylabel('cos(x)', fontsize=12)
ax[1].legend()
ax[1].grid(True)

'''Аннотация для косинуса'''
ax[1].annotate('Максимум', xy=(0, 1), xytext=(1, 0.5),
                arrowprops=dict(facecolor='black', shrink=0.05),
                fontsize=10, color='red')
ax[1].annotate('Минимум', xy=(3.14, -1), xytext=(4, -0.5),
                arrowprops=dict(facecolor='black', shrink=0.05),
                fontsize=10, color='red')

'''Добавление второго Y-оси на первом подграфике'''
ax2 = ax[0].twinx()
ax2.plot(x, y2, color='green', label='cos(x) на sin(x)', linestyle=':', linewidth=1)
ax2.set_ylabel('cos(x)', fontsize=12)
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

### 7. График распределения

```python
'''Генерация случайных данных'''
data = sns.load_dataset('iris')

'''Создание графика распределения'''
sns.histplot(data=data, x='sepal_length', kde=True)
plt.title('Распределение длины чашелистика')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

'''Генерация случайных данных'''
data = sns.load_dataset('iris')

'''Создание графика'''
plt.figure(figsize=(10, 6))

'''Создание графика распределения с KDE для всех видов'''
sns.histplot(data=data, x='sepal_length', hue='species', kde=True,
             multiple='stack', palette='Set1', alpha=0.7)

'''Настройка заголовка и меток'''
plt.title('Распределение длины чашелистика по видам', fontsize=16)
plt.xlabel('Длина чашелистика (см)', fontsize=12)
plt.ylabel('Частота', fontsize=12)

'''Добавление аннотаций для максимальных значений'''
for species in data['species'].unique():
    max_length = data[data['species'] == species]['sepal_length'].max()
    count = data[data['species'] == species]['sepal_length'].value_counts().max()
    plt.annotate(f'Макс. {species}: {max_length:.2f} см',
                 xy=(max_length, count), xytext=(max_length + 0.2, count + 5),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 fontsize=10, color='black')

'''Настройка сетки'''
plt.grid(True)

'''Показать график'''
plt.tight_layout()
plt.show()
```

### 8. Корреляционная матрица и тепловая карта

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

'''Загрузка набора данных Iris'''
data = sns.load_dataset('iris')

'''Вычисление корреляционной матрицы, используя только числовые столбцы'''
correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()

'''Создание тепловой карты'''
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Корреляционная матрица')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

'''Загрузка набора данных Iris'''
data = sns.load_dataset('iris')

'''Вычисление корреляционной матрицы, используя только числовые столбцы'''
correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()

'''Создание маски для верхнего треугольника'''
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

'''Создание тепловой карты'''
plt.figure(figsize=(10, 8))
sns.set(style='white')  #

'''Настройка тепловой карты'''
heatmap = sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm',
                       fmt='.2f', square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})

'''Настройка заголовка'''
plt.title('Корреляционная матрица Iris', fontsize=18)

'''Настройка меток осей'''
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)

'''Установка границ для лучшего восприятия'''
plt.grid(False)
plt.tight_layout()

'''Показать график'''
plt.show()
```

### 9. Парные графики

```python
'''Создание парных графиков'''
sns.pairplot(data, hue='species', markers=["o", "s", "D"])
plt.title('Парные графики для набора данных Iris')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

'''Загрузка набора данных Iris'''
data = sns.load_dataset('iris')

'''Настройка стиля графиков'''
sns.set(style='whitegrid')

'''Создание парных графиков с дополнительными настройками'''
pair_plot = sns.pairplot(data, hue='species', markers=["o", "s", "D"],
                         palette='Set2', diag_kind='kde', height=2.5)

'''Настройка заголовка (применяется к каждому подграфику)'''
pair_plot.fig.suptitle('Парные графики для набора данных Iris', y=1.02, fontsize=18)

'''Добавление аннотаций на диаграммы'''
for i in range(len(data.columns)-1):
    for j in range(len(data.columns)-1):
        if i != j:
            ax = pair_plot.axes[i, j]
            for species in data['species'].unique():
                subset = data[data['species'] == species]
                mean_x = subset[data.columns[j]].mean()
                mean_y = subset[data.columns[i]].mean()
                ax.annotate(f'{species}', xy=(mean_x, mean_y),
                            textcoords='offset points', xytext=(0, 10),
                            ha='center', fontsize=10, color='black', weight='bold')

'''Показать график'''
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

'''Загрузка набора данных Iris'''
data = sns.load_dataset('iris')

'''Настройка стиля графиков'''
sns.set(style='whitegrid')

'''Создание столбчатого графика'''
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='species', y='sepal_length', data=data, palette='pastel', ci='sd')

'''Настройка заголовка и меток'''
plt.title('Средняя длина чашелистика по видам', fontsize=16)
plt.xlabel('Вид', fontsize=12)
plt.ylabel('Средняя длина чашелистика (см)', fontsize=12)

'''Добавление аннотаций для значений'''
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():.2f}',
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='bottom', fontsize=10, color='black',
                      xytext=(0, 5), textcoords='offset points')

'''Добавление сетки'''
plt.grid(axis='y', linestyle='--', alpha=0.7)

'''Показать график'''
plt.tight_layout()
plt.show()
```

### 11. График рассеяния с регрессионной линией

```python
import seaborn as sns
import matplotlib.pyplot as plt

'''Загрузка набора данных Iris'''
data = sns.load_dataset('iris')

'''Настройка стиля графиков'''
sns.set(style='whitegrid')

'''Создание графика рассеяния с регрессионной линией'''
plt.figure(figsize=(10, 6))
scatter_plot = sns.regplot(x='sepal_length', y='sepal_width', data=data,
                            scatter_kws={"s": 100, "alpha": 0.7, "edgecolor": "w"},
                            line_kws={"color": "red", "lw": 2}, ci=95)

'''Настройка заголовка и меток'''
plt.title('График рассеяния с регрессионной линией', fontsize=16)
plt.xlabel('Длина чашелистика (см)', fontsize=12)
plt.ylabel('Ширина чашелистика (см)', fontsize=12)

'''Добавление аннотаций для каждого вида'''
for species in data['species'].unique():
    subset = data[data['species'] == species]
    mean_x = subset['sepal_length'].mean()
    mean_y = subset['sepal_width'].mean()
    plt.annotate(species, xy=(mean_x, mean_y),
                 textcoords='offset points', xytext=(0, 10),
                 ha='center', fontsize=10, color='black', weight='bold')

'''Добавление сетки'''
plt.grid(True)

'''Показать график'''
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

'''Данные'''
x = [1, 2, 3, 4, 5]
y1 = [10, 11, 12, 13, 14]
y2 = [14, 13, 12, 11, 10]
y3 = [12, 14, 13, 15, 16]

'''Создание графика'''
fig = go.Figure()

'''Добавление линий'''
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines+markers', name='Линейный график 1',
                         line=dict(color='blue', width=2), marker=dict(size=8)))

fig.add_trace(go.Scatter(x=x, y=y2, mode='lines+markers', name='Линейный график 2',
                         line=dict(color='red', dash='dash', width=2), marker=dict(size=8)))

fig.add_trace(go.Scatter(x=x, y=y3, mode='lines+markers', name='Линейный график 3',
                         line=dict(color='green', dash='dot', width=2), marker=dict(size=8)))

'''Настройка макета'''
fig.update_layout(
    title='Сложный график с использованием Plotly Graph Objects',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    legend_title='Легенда',
    template='plotly_white',
    hovermode='x unified'
)

'''Добавление аннотации'''
fig.add_annotation(
    x=3,
    y=12,
    text='Среднее значение',
    showarrow=True,
    arrowhead=2,
    ax=0,
    ay=-40,
    font=dict(color='black', size=12)
)

'''Настройка сетки'''
fig.update_xaxes(showgrid=True, gridcolor='LightGray')
fig.update_yaxes(showgrid=True, gridcolor='LightGray')

'''Отображение графика'''
fig.show()
```

```python
import plotly.express as px

'''Загрузка набора данных'''
df = px.data.iris()

'''Создание графика рассеяния с настройками'''
fig = px.scatter(
    df,
    x='sepal_length',
    y='sepal_width',
    color='species',
    size='petal_length',  # Размер точек в зависимости от длины лепестка
    hover_name='species',  # Подсказка с названием вида
    title='График рассеяния Iris с дополнительными параметрами',
    size_max=20,  # Максимальный размер точки
    template='plotly_white'
)

'''Добавление линии регрессии'''
fig.add_trace(px.line(df, x='sepal_length', y='sepal_width', color='species').data[0])

'''Настройка аннотаций'''
mean_length = df['sepal_length'].mean()
mean_width = df['sepal_width'].mean()
fig.add_annotation(
    x=mean_length,
    y=mean_width,
    text="Среднее значение",
    showarrow=True,
    arrowhead=2,
    ax=0,
    ay=-40,
    font=dict(color='black', size=12)
)

'''Настройка макета'''
fig.update_layout(
    xaxis_title='Длина чашелистика (см)',
    yaxis_title='Ширина чашелистика (см)',
    legend_title='Вид',
    hovermode='closest'
)

'''Отображение графика'''
fig.show()
```

```python
import plotly.express as px

'''Загрузка набора данных'''
df = px.data.iris()

'''Создание интерактивной гистограммы'''
fig = px.histogram(
    df,
    x='sepal_length',
    color='species',
    title='Гистограмма длины чашелистика с дополнительными параметрами',
    histnorm='probability density',  # Нормировка по плотности
    barmode='overlay',  # Накладывание гистограмм
    opacity=0.75,  # Прозрачность
    template='plotly_white',
    marginal='rug'  # Добавление рога для отображения распределения
)

'''Настройка осей'''
fig.update_xaxes(title_text='Длина чашелистика (см)', range=[4, 8])
fig.update_yaxes(title_text='Плотность вероятности')

'''Добавление аннотаций'''
fig.add_annotation(
    x=5.5,
    y=0.3,
    text="Максимальная плотность",
    showarrow=True,
    arrowhead=2,
    ax=0,
    ay=-40,
    font=dict(color='black', size=12)
)

'''Настройка легенды'''
fig.update_layout(legend_title_text='Вид', legend=dict(x=0.8, y=1))

'''Отображение графика'''
fig.show()
```

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px

'''Создание приложения Dash'''
app = dash.Dash(__name__)

'''Загрузка данных'''
df = px.data.iris()

'''Макет приложения'''
app.layout = html.Div(children=[
    html.H1(children='Приложение Dash с Plotly'),

    dcc.Dropdown(
        id='species-dropdown',
        options=[{'label': species, 'value': species} for species in df['species'].unique()],
        value='setosa',  # Значение по умолчанию
        multi=False,
        clearable=False,
        style={'width': '50%', 'margin-bottom': '20px'}
    ),

    dcc.Graph(
        id='example-graph'
    )
])

'''Обновление графика на основе выбранного вида'''
@app.callback(
    Output('example-graph', 'figure'),
    Input('species-dropdown', 'value')
)
def update_graph(selected_species):
    filtered_df = df[df['species'] == selected_species]
    fig = px.scatter(
        filtered_df,
        x='sepal_length',
        y='sepal_width',
        color='species',
        title=f'График рассеяния для {selected_species}',
        size='petal_length',  # Размер точек в зависимости от длины лепестка
        hover_name='species'
    )
    return fig

'''Запуск приложения'''
if __name__ == '__main__':
    app.run_server(debug=True)
```

3. Создание интерактивного графика с панелью управления

```python
from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import curdoc

'''Создание фигуры'''
p = figure(title="Интерактивный график", x_axis_label='X-axis', y_axis_label='Y-axis')
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
line = p.line(x, y, line_width=2)

'''Функция для обновления графика'''
def update(attr, old, new):
    scale = slider.value
    line.data_source.data['y'] = [i * scale for i in y]

'''Создание слайдера'''
slider = Slider(start=0.1, end=2, value=1, step=0.1, title="Масштаб")
slider.on_change('value', update)

'''Организация элементов на странице'''
layout = column(slider, p)

'''Добавление макета на страницу'''
curdoc().add_root(layout)
```

5. Создание тепловой карты

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.io import output_notebook

'''Включение вывода в Jupyter Notebook'''
output_notebook()

'''Данные для тепловой карты'''
data = np.random.rand(10, 10)

'''Преобразование данных в нужный формат'''
source = ColumnDataSource(data={'x': np.repeat(np.arange(10), 10),
                                 'y': np.tile(np.arange(10), 10),
                                 'value': data.flatten()})

'''Создание фигуры'''
p = figure(title="Тепловая карта", x_axis_label='X-axis', y_axis_label='Y-axis',
           tools="hover", tooltips=[("Value", "@value")], width=400, height=400)

'''Определение цвета'''
mapper = linear_cmap(field_name='value', palette='Viridis256', low=0, high=1)

'''Добавление квадратиков'''
p.rect(x='x', y='y', width=1, height=1, source=source, fill_color=mapper)

'''Добавление цветовой шкалы'''
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

'''Отображение графика'''
show(p)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, Slider, CustomJS
from bokeh.layouts import column

'''Включение вывода в Jupyter Notebook'''
output_notebook()

'''Данные'''
x = [1, 2, 3, 4, 5]
y1 = [6, 7, 2, 4, 5]
y2 = [5, 6, 3, 5, 7]

'''Создание источника данных'''
source = ColumnDataSource(data=dict(x=x, y1=y1, y2=y2))

'''Создание фигуры'''
p = figure(title="Интерактивный линейный график", x_axis_label='X-axis', y_axis_label='Y-axis')

'''Добавление линий'''
p.line('x', 'y1', source=source, legend_label='Линия 1', line_width=2, color='blue')
p.line('x', 'y2', source=source, legend_label='Линия 2', line_width=2, color='red')

'''Настройка сетки'''
p.grid.grid_line_color = 'lightgray'
p.grid.grid_line_alpha = 0.7

'''Добавление аннотаций'''
p.text(x=[1, 5], y=[6, 7], text=["Линия 1", "Линия 2"], text_color=['blue', 'red'],
        text_font_size='10pt', text_align='center')

'''Создание слайдера для изменения данных'''
slider = Slider(start=0, end=10, value=1, step=1, title="Изменение Y-значений")

'''Обновление данных при изменении слайдера'''
callback = CustomJS(args={'source': source}, code="""
    const data = source.data;
    const y1 = data['y1'];
    const y2 = data['y2'];
    const value = cb_obj.value;

    for (let i = 0; i < y1.length; i++) {
        y1[i] = y1[i] + value;
        y2[i] = y2[i] + value;
    }
    source.change.emit();
""")

slider.js_on_change('value', callback)

'''Компоновка графика и слайдера'''
layout = column(p, slider)

'''Отображение графика'''
show(layout)
```

```python
import numpy as np
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, ColorPicker, CheckboxGroup, Slider, Button, CustomJS
from bokeh.layouts import row, column
from bokeh.palettes import Category10_8
from bokeh.events import DoubleTap

# Включаем вывод графики прямо в блокнот
output_notebook()

# Генерируем данные
N = 100
x = np.linspace(0, 10, N)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x / 2)

# Создаем источник данных
source = ColumnDataSource(data=dict(x=x, y1=y1, y2=y2, y3=y3))

# Цветовая палитра
colors = Category10_8[:3]

# Создание фигуры
p = figure(
    title="Интерактивная динамика функций sin/cos/tan",
    height=400,  # Исправленный аргумент высоты
    width=800,   # Исправленный аргумент ширины
    tools="pan,wheel_zoom,reset,save",
    toolbar_location="above"
)

# Добавляем линии на графике
line1 = p.line('x', 'y1', source=source, legend_label="sin(x)", line_width=2, color=colors[0])
line2 = p.line('x', 'y2', source=source, legend_label="cos(x)", line_width=2, color=colors[1])
line3 = p.line('x', 'y3', source=source, legend_label="tan(x/2)", line_width=2, color=colors[2])

# Настройки легенды
p.legend.location = "top_left"
p.legend.click_policy = "hide"

# Добавляем точки на пересечении осей координат
circle = p.circle([0], [0], size=10, fill_color="black")

# Настройка фона и сеток
p.background_fill_color = "#fafafa"
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Аннотации
p.text(x=[0], y=[0], text=["Ось пересечения"], text_color="black", text_font_size="10pt", text_align="right")

# Элементы управления цветом линий
color_picker_sin = ColorPicker(color=colors[0], width=100, height=30, title="Цвет sin(x)")
color_picker_cos = ColorPicker(color=colors[1], width=100, height=30, title="Цвет cos(x)")
color_picker_tan = ColorPicker(color=colors[2], width=100, height=30, title="Цвет tan(x/2)")

# Связываем выбор цвета с линиями
line1.glyph.update(line_color=color_picker_sin.color)
line2.glyph.update(line_color=color_picker_cos.color)
line3.glyph.update(line_color=color_picker_tan.color)

# Обработчики выбора цвета
def update_color(attr, old, new):
    if attr == 'color':
        line1.glyph.update(line_color=color_picker_sin.color)
        line2.glyph.update(line_color=color_picker_cos.color)
        line3.glyph.update(line_color=color_picker_tan.color)

color_picker_sin.on_change("color", update_color)
color_picker_cos.on_change("color", update_color)
color_picker_tan.on_change("color", update_color)

# Контроль видимости линий
checkbox_group = CheckboxGroup(labels=["Показать sin(x)", "Показать cos(x)", "Показать tan(x/2)"],
                               active=[0, 1, 2],
                               inline=True)

# Обработка события смены чекбоксов
def checkbox_callback(attr, old, new):
    if 0 in checkbox_group.active:
        line1.visible = True
    else:
        line1.visible = False

    if 1 in checkbox_group.active:
        line2.visible = True
    else:
        line2.visible = False

    if 2 in checkbox_group.active:
        line3.visible = True
    else:
        line3.visible = False

checkbox_group.on_change('active', checkbox_callback)

# Анимационный слайдер для изменения масштаба графика
slider = Slider(start=-10, end=10, value=0, step=0.1, title="Масштаб графика")

js_code_slider = """
const scaleFactor = Math.pow(cb_obj.value, 2);
for (var key in source.data) {
    if (key.startsWith('y')) {
        var y = source.data[key];
        for (var i = 0; i < y.length; i++) {
            y[i] *= scaleFactor;
        }
    }
}
source.change.emit();
"""

slider.js_on_change('value', CustomJS(args=dict(source=source), code=js_code_slider))

# Кнопка сброса значений
button_reset = Button(label="Сбросить значения", button_type="primary")

js_code_button = """
source.data['y1'] = original_data['y1'].slice(); // Используем slice() для копирования массива
source.data['y2'] = original_data['y2'].slice();
source.data['y3'] = original_data['y3'].slice();
source.change.emit();
"""

original_data = dict(y1=y1.copy(), y2=y2.copy(), y3=y3.copy())
button_reset.js_on_click(CustomJS(args=dict(source=source, original_data=original_data), code=js_code_button))

# Дополнительные эффекты при двойном клике на графику
def doubletap_handler(event):
    print(f"Двойной клик произошел в точке ({event.x}, {event.y})")

p.on_event(DoubleTap, doubletap_handler)

# Компоновка элементов интерфейса
controls = column(slider, color_picker_sin, color_picker_cos, color_picker_tan, checkbox_group, button_reset)
layout = row(controls, p)

# Отображаем интерактивный график
show(layout)
```
