```python
import pandas as pd
import os
from datetime import datetime
```

```python
def merge_and_save_to_csv():
    """–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ 4 Excel —Ñ–∞–π–ª–æ–≤ –≤ –æ–¥–∏–Ω CSV (—Ç–æ–ª—å–∫–æ —á—Ç–µ–Ω–∏–µ –∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ)"""
    
    # –ü—É—Ç–∏ –∫ —Ñ–∞–π–ª–∞–º –≤ –ø–∞–ø–∫–µ data
    FILE_PATHS = {
        'customer': 'Customer.xlsx',
        'product': 'Product.xlsx', 
        'sales': 'Sales.xlsx',
        'territories': 'Territories.xlsx'
    }
    
    print("üîç –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ñ–∞–π–ª–æ–≤...")
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –≤—Å–µ—Ö —Ñ–∞–π–ª–æ–≤
    for file_type, file_path in FILE_PATHS.items():
        if os.path.exists(file_path):
            print(f"‚úÖ {file_type}: {file_path} - –Ω–∞–π–¥–µ–Ω")
        else:
            print(f"‚ùå {file_type}: {file_path} - –Ω–µ –Ω–∞–π–¥–µ–Ω")
            return None
    
    try:
        # –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
        print("\nüì• –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö...")
        customer_df = pd.read_excel(FILE_PATHS['customer'])
        product_df = pd.read_excel(FILE_PATHS['product'])
        sales_df = pd.read_excel(FILE_PATHS['sales'])
        territories_df = pd.read_excel(FILE_PATHS['territories'])
        
        print("‚úÖ –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
        print(f"üìä –†–∞–∑–º–µ—Ä—ã –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
        print(f"   - Customers: {customer_df.shape}")
        print(f"   - Products: {product_df.shape}")
        print(f"   - Sales: {sales_df.shape}")
        print(f"   - Territories: {territories_df.shape}")
        
        # –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö
        print("\nüîÑ –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö...")
        
        # 1. –û–±—ä–µ–¥–∏–Ω—è–µ–º Sales —Å Customer
        merged_df = sales_df.merge(
            customer_df, 
            on='CustomerKey', 
            how='left', 
            suffixes=('', '_customer')
        )
        print(f"   ‚úÖ Sales + Customers: {merged_df.shape}")
        
        # 2. –î–æ–±–∞–≤–ª—è–µ–º Product –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
        merged_df = merged_df.merge(
            product_df, 
            on='ProductKey', 
            how='left', 
            suffixes=('', '_product')
        )
        print(f"   ‚úÖ + Products: {merged_df.shape}")
        
        # 3. –î–æ–±–∞–≤–ª—è–µ–º Territories –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é
        merged_df = merged_df.merge(
            territories_df, 
            left_on='SalesTerritoryKey', 
            right_on='SalesTerritoryKey',
            how='left', 
            suffixes=('', '_territory')
        )
        print(f"   ‚úÖ + Territories: {merged_df.shape}")
        
        # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ CSV
        print("\nüíæ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤ CSV...")
        output_filename = "merged_sales_data.csv"
        merged_df.to_csv(output_filename, index=False, encoding='utf-8')
        
        print("‚úÖ –ì–æ—Ç–æ–≤–æ!")
        print(f"üìä –§–∏–Ω–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞: {merged_df.shape}")
        print(f"üíæ –§–∞–π–ª —Å–æ—Ö—Ä–∞–Ω–µ–Ω –∫–∞–∫: {output_filename}")
        
        # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫–æ–ª–æ–Ω–∫–∞—Ö
        print(f"üìã –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–ª–æ–Ω–æ–∫: {len(merged_df.columns)}")
        print("\nüìù –°–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –∫–æ–ª–æ–Ω–æ–∫:")
        for i, col in enumerate(merged_df.columns, 1):
            print(f"   {i:2d}. {col}")
        
        return merged_df
        
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
        return None

# –ó–∞–ø—É—Å–∫ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è
final_df = merge_and_save_to_csv()
```
